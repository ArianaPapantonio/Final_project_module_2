## Objetivo : Generar un range que busque las 500 canciones de un género 

```python
#Objetivo : sacar las primeras 500 canciones del genero genre:electronic year:2023 con nombre del artista y su cancion usando range en el offset

electronica_2021 = {"nombre_cancion_album": [], "artista": [], "tipo_album_cancion" : [], "año_lanzamiento" : [], "género" : []} ##Ver contador 

for ind in range(0,501,50):
    
    datos = sp.search(q="genre:electronica year:2021", type='track', limit=50, offset=ind)

    json_sp = datos["tracks"]["items"]

    for elemento in json_sp:
        electronica_2021["nombre_cancion_album"].append(elemento["name"]) #Añade nombre cancion
        electronica_2021["tipo_album_cancion"].append(elemento["album"]["album_type"]) #Añade si es album o single
        release_date = elemento["album"]["release_date"] #Define fecha de lanzamiento
        year = release_date.split("-")[0] #Divide la fecha
        electronica_2021["año_lanzamiento"].append(year) #Append solo el año de lanzamiento
        electronica_2021["artista"].append(elemento["album"]["artists"][0]["name"]) #Añade nombre de artista (solo uno por cancion)
        electronica_2021["género"].append("Electrónica") #Añade género
    
```

```python
for key, value in electronica_2021.items():
    print(f"{key}: {len(value)}")
```

```python
electronica_2021 = pd.DataFrame(electronica_2021) 
electronica_2021
```
