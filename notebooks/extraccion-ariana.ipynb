```python
#Electrónica 2021: 301 - 350

datos = sp.search(q="genre:electronica year:2021", type='track', limit=50, offset=301)

json_sp = datos["tracks"]["items"]

# Este es el dict donde queremos almacenar la info
dic_canciones1 = {"song_name": [], "artist_name": [], "type" : [], "release_date" : []}

for elemento in json_sp:
    dic_canciones1["song_name"].append(elemento["name"]) #Añade nombre cancion
    dic_canciones1["type"].append(elemento["album"]["album_type"]) #Añade si es album o single
    release_date = elemento['album']["release_date"] #Define fecha de lanzamiento
    year = release_date.split("-")[0] #Divide la fecha
    dic_canciones1["release_date"].append(year) #Append solo el año de lanzamiento

    for dato in elemento["album"]["artists"]: 
        dic_canciones1["artist_name"].append(dato["name"]) #Añade nombre de artista 
    
    
           
```
